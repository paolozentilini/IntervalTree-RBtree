```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

time = np.loadtxt("CodiceVersione2/Results.dat")
time /= 1e6
x = np.arange(1,501,1)

def func(x,a):
    return a * np.log2(x)

alpha, cov = curve_fit(func, x, time)
alpha
```

```python
plt.figure(figsize=(12,6))

plt.title('Time (# nodes) ',fontsize=20)
plt.plot(x, time, color='black',label='measures',lw=1)
plt.plot(x, alpha*np.log2(x),color='r',label=r'$y\,\simeq\,$'+str(round(alpha[0],2))+'$\,\cdot\,\log_2(x)\,$')
plt.legend(fontsize=16)
plt.grid(True)
plt.xlabel('# nodes',fontsize=18)
plt.ylabel('Time [ns]',fontsize=18)
#plt.savefig('Time.pdf')
plt.show()
```
